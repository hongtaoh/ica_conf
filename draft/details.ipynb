```python
import pandas as pd 
import numpy as np 
import json 
```

```python
papers_path = '../data/api/papers.json'
authors_path = '../data/api/authors.json'
sessions_path = '../data/api/sessions.json'
```

```python
with open(papers_path, 'r') as file:
    papers = json.load(file) 

with open(authors_path, 'r') as file:
    authors = json.load(file) 

with open(sessions_path, 'r') as file:
    sessions = json.load(file) 
```

```python
len(papers), len(authors), len(sessions)
```

```python
import requests

base_url = "http://127.0.0.1:8000"

# Parameters for the request
params = {
    "paper_count" : 10
}

# Make the request
response = requests.get(f"{base_url}/sessions", params=params)

# print("Request URL:", response.url)

# Check the response
if response.status_code == 200:
    papers = response.json()
    print("Papers retrieved:", papers)
else:
    print("Failed to retrieve papers:", response.status_code, response.text)
```
